# Fairall 9 Photometry

Minimal example: resolve a target with `SkyCoord.from_name`, query Bandwagon's default catalogs, and convert the raw XMatch results into long-form photometry.


```python
from pathlib import Path
import sys

from astropy.coordinates import SkyCoord

src_root = Path.cwd().resolve().parent / "src"
if src_root.is_dir() and str(src_root) not in sys.path:
    sys.path.insert(0, str(src_root))

from bandwagon import matches_to_photometry, xmatch_catalogs
```

```python
coord = SkyCoord.from_name("Fairall 9")
coord
```

```python
matches = xmatch_catalogs(coord, source_id=["Fairall 9"])

for name, table in matches.items():
    print(f"{name}: {len(table)} matched rows")
```

```python
photometry = matches_to_photometry(matches, max_mag_err=1.0)
photometry["source_id", "catalog", "filter_name", "flux_mjy", "flux_err_mjy", "psf_fwhm_arcsec", "match_distance_arcsec"]
```
